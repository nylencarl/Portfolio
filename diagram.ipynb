```python
# %%
# Importera bibliotek
import pandas as pd
import matplotlib.pyplot as plt

# Gör så att grafer visas snyggt i notebooken
%matplotlib inline
```

```python
# %%
# Skapa en liten exempel-dataset
data = {
    "Månad": ["Jan", "Feb", "Mar", "Apr", "Maj", "Jun"],
    "Försäljning": [120, 150, 170, 200, 230, 250],
    "Kunder": [30, 35, 40, 50, 60, 70]
}

df = pd.DataFrame(data)

# Visa datan
df
```

```python
# %%
# Enkel statistik
print("Medelförsäljning:", df["Försäljning"].mean())
print("Max försäljning:", df["Försäljning"].max())
print("Antal månader:", len(df))
```

```python
# %%
# Rita en linjegraf
plt.plot(df["Månad"], df["Försäljning"], marker="o", label="Försäljning")
plt.plot(df["Månad"], df["Kunder"], marker="s", label="Kunder")

plt.title("Försäljning och antal kunder över tid")
plt.xlabel("Månad")
plt.ylabel("Antal")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# %%
# Rita ett stapeldiagram
df.plot(kind="bar", x="Månad", y="Försäljning", legend=False)

plt.title("Försäljning per månad")
plt.ylabel("Försäljning")
plt.show()
```
