### 2. Python Koncept: Variabler

### Variabler
- En variabel i Python är ett namn som används för att lagra och referera till ett värde i datorns minne

```python
name = "Jonathan"

city = "Mölndal"

print(name, " bor i ", city, "\nnyrad")
```

##### Variabler med f-string (formaterade strängar)
Vi använder f-string när vi vill enkelt printa variabler direkt i en sträng. Du skriver ett f framför strängen och placerar variabeln inom {}


```python
print(f"{name} bor i {city}Sverige")
```

```python
print(f"""
      Jonathan bor i {city}
      Sverige
""")
```

```python
beskrivning = f"""
      {name} bor i {city}
      Sverige
      sadfsadf
      sadfasdf
      asdfasfdasf
      sadfasfdas
"""
```

Vi kan även använda expressions direkt i vår f-string

```python
print(f"{name} bor i {city.upper()}")  # upper() omvandlar city till versaler
```

```python
city = "LULEÅ"
print(city)
```

Välja ut delar av en sträng (slicing)

```python
print(f"Första 5 bokstäverna i name är {name[0:3].upper()}{name[3:].lower()}")
print(f"Sista 3 bokstäverna i city är {city[-3:]}")
```

#### Variabler vs Konstanter i Python

I Python finns inte ett inbyggt sätt att skapa riktiga konstanter (som i vissa andra språk), men det finns en konvention:

Om en variabel skrivs med VERSALER betraktas den som en konstant, dvs. den ska inte ändras.

Observera att det tekniskt sett går att ändra en konstant variabel i Python men konventionen är att den inte ska ändras för bättre läsbarhet och kodunderhåll.

```python
PI = 3.14 # Detta är en konstant, vi bör inte ändra den
print(PI)
```

Exemplet nedan demonstrerar användningsområdet för konstanter.
I detta fall är skattesatsen i procent ett fast värde som inte kommer ändras, men pris på olika produkter varierar ofta därav variabeln.

```python
SKATTESATS = 0.25

pris_exkl = 150

moms = pris_exkl * SKATTESATS

pris_inkl = pris_exkl + moms

print(f"Pris exkl. moms: {pris_exkl}")
print(f"Moms: {moms}")
print(f"Pris inkl. moms: {pris_inkl}")
```

### Python koncept: Datatyper, Strukturer

Likt andra programmeringsspråk har Python ett gäng inbyggda datatyper. Men till skillnad från språk som Java behöver vi i Python inte tala om vilken datatyp våra variabler ska hålla. Fördelen är att vi sparar mycket på kod men största nackdelen är att vi kan få in oväntade datatyper i våra variabler vilket kan orsaka errors.

Exempel: Vi säger att x = 10 och sedan delar vi x med 2.3 ersätter x med det värdet som plötsligt har blivit ett decimaltal. Om vårat program förväntar sig ett heltal vid något tillfälle kan vi få problem.

Integer (int) - Heltal t.ex. 1, -7, 42 <br>
Float (float) - Flytande decimaltal 

```python
# INTEGERS
x = 10
y = 5

print(f"x = {x}")
print(f"y = {y}")
print(f"{x + y}")

print(f"Vilken datatyp är x? Vi använder type() för att ta reda på det: x = {type(x)}")
```

```python
# FLOATS
x = 5-0.3

print(f"Vi har nu skrivit över x med ett float-värde: x = {x}")
print(f"Vilken datatyp är x nu? {type(x)}")
```

String (str)

List
- En ordnad samling av element. Du kan ändra, lägga till, ta bort, sortera element i en lista
- Användningsområde? När du vill samla flera värden i en ordnad följd, t.ex. en lista med kolumnnamn eller tabeller


```python
tables = ["users", "orders", "products"]

columns = ["id", "name", "email", "price", "quantity", "created_time", "updated_time"]

print(f"tables: {type(tables)}")
print(f"columns: {type(columns)}")
```

```python
print(f"Alla element i listan columns: {columns}")
```

```python
print(f"Första kolumnen i listan columns: {columns[0:3]}")
print(f"Sista kolumnen i listan columns: {columns[-1]}")
```

```python
# Utforksa listorna

# Skriv ut längden på lista (i detta fall antalet element i listan)
print(f"Antal element i listan columns: {len(columns)}")
```

```python
columns_to_drop = ["created_time", "updated_time"]

for col in columns_to_drop: # Loopar igenom varje kolumn i columns_to_drop och tar bort den från columns
    columns.remove(col)
print(f"Återstående kolumner efter borttagning: {columns}")
```

```python
print(columns)
```

```python
full_name = "full_name"

columns.append(full_name) # Lägger till kolumnen full_name i listan columns
print(f"Kolumner efter full_name har lagts till: {columns}")
```

Tuple
- En ordnad samling som inte kan ändras. Används för fasta grupper av värden.

```python
inte_platt = ("Jorden", "Månen", "Saturnus")

print(f"inte_platt datatyp: {type(inte_platt)}")
print(f"Det tredje elementet i inte_platt: {inte_platt[2]}")
```

Dictionary
- En osorterad samling av nyckel-värde par (key-value pairs). Du slår upp värden med nycklar.
- Användingsområde? När du vill lagra strukturerad data, t.ex. en persons uppgifter.

```python
person = {
    "name": "Jonathan",
    "age": 35,  # None används för att representera avsaknad av värde
    "city": "Luleå",
    "is_alive": True
    }

print(f"person dict datatype: {type(person)}")
```

```python
print(f"Personens namn: {person['name']}")
print(f"Personens ålder: {person['age']}")
```

```python
del person["age"]  # Tar bort ålder från person
print(f"Efter att ha tagit bort ålder kolumnenr: {person}")
```

8. Set
- En osorterad samling av unika värden.
- Användningsområde? När du vill veta vilka unika värden som finns, eller jämföra två samlingar (t.ex. elever som lämnat in vs inte lämnat in en uppgift)

```python
namn = ["Jonathan", "Jonathan", "Anna", "Mattias"]
print(namn)
```

```python
unika_namn = set(namn)
print(unika_namn)
```

```python
print(f"Antal namn: {len(namn)}")
print(f"Antal unika namn: {len(unika_namn)}")
```
