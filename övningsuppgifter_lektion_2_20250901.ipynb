4. Separera en sträng till flera strängar
- Från **email** variabeln ska du skapa en ny variabel som heter **full_name**
- Hint1: Ta reda på hur man väljer ut delar av en sträng (slicing)
- Hint2: Ta reda på hur man gör första bokstaven i namnet till en stor bokstav
- Printa ut din nya variabel **full_name** som bör visa "Johan Karlsson"

```python
email = "johan.karlsson@företag.se"

# Fortsätt här
email = "johan.karlsson@företag.se"

# 1. Ta ut delen före @
name_part = email.split("@")[0]   # -> "johan.karlsson"

# 2. Separera förnamn och efternamn
first, last = name_part.split(".")   # -> "johan", "karlsson"

# 3. Gör första bokstaven stor
first = first.capitalize()           # -> "Johan"
last = last.capitalize()             # -> "Karlsson"

# 4. Sätt ihop till full_name
full_name = f"{first} {last}"

print(full_name)         # -> Johan Karlsson
print(type(full_name))   # -> <class 'str'>

#Det går också ta .title() istället för .capitalize()
```

5. Lista
- Skapa en lista som innehåller 5 företagsnamn
- Efter det ska du droppa det fjärde företaget från listan
- Printa slutgiltiga listan
- Printa även längden av listan (antalet element i listan)

```python
companies = ["ICA", "ALDI", "Hemköp", "Coop", "Willys"]

# Ta bort fjärde företaget (index 3)
companies_to_drop = [companies[3]]

for com in companies_to_drop:
    companies.remove(com)

print(f"Återstående företag efter borttagning: {companies}")
```
