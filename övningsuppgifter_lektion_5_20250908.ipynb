```python
Nullvärden = df["total_bedrooms"].isnull().sum()
antal_rader = df.count()
tot = len(df)
sängar = len(df["total_bedrooms"])

total = (tot - Nullvärden) / (tot)
print(total)
```

8. Du bör ha kommit fram till att det är en liten procentandel värden som saknas i kolumnen.

Din första tanke är nog 'hur hanterar jag detta?'
- Du ställer frågan till mottagaren av rapporten och förklarar situationen, personen säger att det går bra att fylla dessa "null" värden med medianen av de värden som finns i kolumnen.
- Uppgiften blir alltså att räkna ut medianen och fylla alla null rader med medianen, detta kan du lösa med enbart Pandas-funktioner.
- Tänk på att detta är transformation som ska fortsätta existera vidare i ditt DF, du behöver alltså uppdatera ditt DF med denna transformation.

```python
median = df["total_bedrooms"].median()
df["total_bedrooms"] = df["total_bedrooms"].fillna(median)
print(df)
```

9. Gör en group by på kolumnen **ocean_proximity** och räkna medellön (**median_income**) per kategori. Sortera högst till lägst.

```python
# Group by och beräkna medelinkomst
resultat = (
    df.groupby("ocean_proximity")["median_income"]
    .mean()                      # medelvärde per kategori
    .sort_values(ascending=False) # sortera högst till lägst
)

print(resultat)
```

13. Visaulisera **median_house_value** per **ocean_proximity** förslagsvis med en bar chart (prova vertikal och horisontell bar chart, vilken tycker du förmedlar informationen bäst?)

Tips: Gör en Group By på ocean_proximity och aggregera medelvärdet på median_house_value

```python
# Gruppera på ocean_proximity och ta medelvärdet av median_house_value
medelvärden = (
    df.groupby("ocean_proximity")["median_house_value"]
    .mean()
    .sort_values(ascending=False)   # sortera så diagrammet blir tydligare
)

print(medelvärden)


```

```python
import matplotlib.pyplot as plt

# Vertikal bar chart
plt.figure(figsize=(8,5))
medelvärden.plot(kind="bar", color="skyblue", edgecolor="black")

plt.title("Genomsnittligt median house value per ocean_proximity")
plt.xlabel("Ocean proximity")
plt.ylabel("Medelvärde av median_house_value")
plt.xticks(rotation=45)
plt.show()
```

```python
# Horisontell bar chart
plt.figure(figsize=(8,5))
medelvärden.plot(kind="barh", color="lightgreen", edgecolor="black")

plt.title("Genomsnittligt median house value per ocean_proximity")
plt.xlabel("Medelvärde av median_house_value")
plt.ylabel("Ocean proximity")
plt.show()
```

14. Visualisera geografisk fördelning av huspriser i Kalifornien
- Använd en scatterplot
- Använd kolumnerna **longitude** och **latitude** för x respektive y axeln
- Hitta ett sätt att färgläggga dina scatter "bollar" baserat på värden i kolumnen **median_house_value**
- Använd en "colormap" för att förtydliga föregående steg

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl

# Exempelkoordinater
sf_coords = (-122.4194, 37.7749)
la_coords = (-118.2437, 34.0522)

plt.figure(figsize=(10,7))

# Skapa Axes
ax = plt.gca()  # hämta nuvarande Axes

# Scatterplot med Seaborn
sns.scatterplot(
    data=df,
    x="longitude",
    y="latitude",
    hue="median_house_value",
    palette="inferno",
    alpha=0.6,
    size=0.5,
    legend=False,
    ax=ax
)

# Lägg till colorbar på samma Axes
norm = mpl.colors.Normalize(vmin=df["median_house_value"].min(), vmax=df["median_house_value"].max())
sm = mpl.cm.ScalarMappable(cmap="inferno", norm=norm)
sm.set_array([])

cbar = plt.colorbar(sm, ax=ax)
cbar.set_label("Median House Value")

# Markera San Francisco och Los Angeles
ax.scatter(*sf_coords, color="red", s=100, label="San Francisco")
ax.scatter(*la_coords, color="blue", s=100, label="Los Angeles")
ax.legend()

# Rubriker
ax.set_title("Geografisk fördelning av huspriser i Kalifornien")
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")

plt.show()
```
