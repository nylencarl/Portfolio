#### Loopar
Det finns två huvudtyper av loopar i Python:
- for-loop - när du vet hur många gånger du ska loopa, till exempel över en lista med kolumner
- while-loop - när du loopar tills ett villkor inte längre är sant, till exempel radera alla värden från en listan tills den är tom.

```python
# vi ska använda oss av time.sleep() för att bättre visualisera looparna
import time 
```

for-loop:

```python
lista = ["ett", "två", "tre", "fyra", "fem"]

# loopa över varje värde i listan (tal representerar nuvarande element)
for tal in lista:
    print(tal)
```

Range: Alternativt loopa över ett antal siffror genom att använda Pythons inbyggda range()

```python
for i in range(5):  # Loopar från 0 till 4
    print(f"Nummer: {i}")
```

Range: Loopa över längden av en lista

```python
lista_hundar = ["Bonus", "Maya", "Alva", "Kiwi"]

for hund in range(len(lista_hundar)): # vi sätter in att range ska vara längden av våran lista med hundnamn
    print(f"Hund {hund}: {lista_hundar[hund]}")
```

enumerate: Istället för att använda range() och skriva in ett värde kan vi istället använda enumerate() med våran lista som input så får vi både index och värde på köpet <br>

```python
lista_städer = ["Stockholm", "Göteborg", "Malmö", "Mölndal", "Luleå"]

# i = index 
# stad = värde från listan
for i, stad in enumerate(lista_städer): # enumerate ger både index och värde direkt och du slipper range(len())
    print(f"Index: {i}  Stad: {stad}")
```

while-loop: <br>
Fortsätter att loopa till villkoret inte längre uppfylls.

```python
# först startar vi en räknare som börjar på 0
räknare = 0

# medan räknaren är mindre än 6 loopar vi, slutar när villkoret inte längre är sant
while räknare < 6:
    print(räknare)
    räknare += 1  # Öka räknaren med 1 varje varv
    time.sleep(0.25)
```

Exempel på hur vi kan använda loops för att skapa och radera filer i vårat repository

importera `os` som vi kan använda till att hantera filer

Använd `os` till att skapa en ny mapp i ditt repo

```python
mapp_namn = "demo_filer"

os.mkdir(mapp_namn)
print(f"Mapp {mapp_namn} skapad.")
```

Skapa ett gäng med Python-filer med en for-loop

```python
antal_filer = 6

for i in range(antal_filer):
    filnamn = f"{mapp_namn}/fil_{i}.py"
    with open(filnamn, "w") as f:
        f.write(f"print({i})")

    print(f"Skapade: {filnamn}")
```

```python
räknare = 0

while räknare < len(filer):
    filnamn = f"{mapp_namn}/{filer[räknare]}"

    os.remove(filnamn)
    time.sleep(1)
    print(f"Raderade: {filnamn}")
    räknare += 1
```

#### If-satser
En if-sats låter oss styra flödet i programmet beroende på ett villkor.

Ett första exempel kan vara att en simpel check om en person får handla på Systembolaget eller inte

```python
ok = "Du får handla på Systembolaget"
inte_ok = "Du får INTE handla på Systembolaget"
nej_men_snart = "Du får INTE handla på Systembolaget. Vänta till nästa födelsedag..."


ålder = 45


if ålder >= 20:
    print(ok)
elif ålder == 19:
    print(nej_men_snart)
else:
    print(inte_ok)
```

Flera villkor

```python
ålder = 25
legitimation = False

if ålder >= 20 and legitimation == True: # du behöver inte egentligen specifiera ""== True" men det är bra för läsbarheten.
    print("Ålder + legitimation OK")
else:
    print("Nekad")
```

Nested

```python
ålder = 25
legitimation = True

if ålder >= 20:
    if legitimation:
        print("Ålder + legitimation OK")
    else:
        print("Nekad på grund av saknad legitimation")
else:
    print("Nekad på grund av ålder")
```

Ternary

```python
ålder = 19
meddelande = "Du får handla på Systembolaget" if ålder >= 20 else "Du får INTE handla på Systembolaget"

print(meddelande)
```

Deklarera variabler i en if-sats

```python
ålder = 21

if ålder >= 20:
    åldern_inne = True
else:
    åldern_inne = False

print(åldern_inne)
```

Ett exempel kan vara att radera mappen vi tidigare skapade som heter **demo_filer**

#### Kombinera funktion, if-sats och loop

```python
def skapa_demo_filer(mapp_namn, antal_filer):
    """
    Denna funktion skapar en mapp och ett antal textfiler i den.\n
    Input:
    - mapp_namn: Namnet på mappen som ska skapas
    - antal: Antalet filer som ska skapas
    """
    # Ovan är en sk. docstring som beskriver funktionen (vi kan hovra över den för att se beskrivningen)

    # Först skapar vi mappen om den inte redan finns
    if not os.path.exists(mapp_namn):
        os.mkdir(mapp_namn)
        proceed = True
    else:
        proceed = False


    # Sedan skapar vi filerna med hjälp av en for-loop
    if proceed == True:
        print(f"SUCCÉ: Mappen {mapp_namn} skapades")
        for i in range(antal_filer):
            filnamn = f"{mapp_namn}/fil_{i}.txt"
            with open(filnamn, "w") as f:
                f.write(f"Innehåll i fil {i}")
                print(f"Skapade: {filnamn}")
                time.sleep(0.25)
    elif proceed == False:
        print("AVBRUTEN: Inga filer skapades eftersom mappen redan fanns")
```

Nu provar vi köra våran funktion genom att kalla på den

```python
skapa_demo_filer(mapp_namn="demo_filer", antal_filer=10)
```

Funktion som raderar våra demo filer

```python
def radera_demo_filer(mapp_namn):
    # Först kollar vi om mappen finns och skapar en lista över filerna samt starter en räknare
    if os.path.exists(mapp_namn):
        filer = os.listdir(mapp_namn)
        räknare = 0

        # Sedan loopar vi igenom filerna och tar bort dem en efter en tills det inte längre finns några filer kvar
        while räknare < len(filer):
            filväg = os.path.join(mapp_namn, filer[räknare])
            os.remove(filväg)
            print(f"Raderade: {filer[räknare]}")
            räknare += 1
            time.sleep(0.25)

    # Slutligen tar vi bort mappen
    try:
        print("Försöker ta bort mappen...")
        time.sleep(2)
        os.rmdir(mapp_namn)
        print(f"Mappen {mapp_namn} har raderats.")
    except Exception as e:
        print(f"Kunde inte radera mappen {mapp_namn}: {e}")
```

```python
radera_demo_filer(mapp_namn="demo_filer")
```
