7. Om du lyckades i föregående steg så ser du att vi har nulls i enbart en kolumn. Vad är omfattningen jämfört med totalt antal rader i DF?
- Räkna ut hur stor procentandel i kolumnen som saknar värden jämfört med totalt antal rader i datasettet.
- Denna information är viktig när vi ska ska resonera kring hanteringen av dessa saknade värden.

```python
column_to_check = "total_bedrooms"

null_count = df.isnull().sum()[column_to_check]
total_rows = df.shape[0]

percentage_nulls = round((null_count / total_rows) * 100, 2)


print(f"{percentage_nulls}% av raderna i kolumnen {column_to_check} saknar värden")
```

8. Du bör ha kommit fram till att det är en liten procentandel värden som saknas i kolumnen.

```python
df["total_bedrooms"] = df["total_bedrooms"].fillna(df["total_bedrooms"].median())
```

12. Varför tror du att vi gjorde en logaritmisk skala? Vad blev det för skillnad från vårat första histogram?
- Svara gärna i en snyggt formaterad Markdown-cell

### Svar: Varför använda logaritmisk skala?

1. **Krymper stora värden mer än små**  
   Exempel (log10): 10 -> 1, 100 -> 2, 1000 -> 3.  
   Skillnaden mellan 10k och 20k blir lika “stor” som mellan 100k och 200k.  

2. **Gör skev data mer balanserad**  
   Många variabler som inkomst och huspriser är "right-skewed" alltså höger-skev.  
   Log-transformen gör fördelningen mer symmetrisk och minskar effekten av outliers.  

3. **Förtydligar mönster i visualiseringar**  
   Små och stora värden hamnar på en skala där båda syns tydligare, vilket gör det lättare att upptäcka samband.  

**Summering:**  
En log-skala hjälper oss hantera stora variationer i data och gör analyser och visualiseringar mer meningsfulla.
