#### Skapa nya kolumner för att analysera lön och bonus
- Vi vill ta fram top 5 och bottom 5 löner/bonusar på employee nivå <br>
- Vi börjar med att normalisera lön och bonus (värde mellan 0 och 1) <br>
- Sedan kombinerar vi till en **score** som är viktat (Lön väger 1 och Bonus väger 2) <br>

````python
# Normalisera lön och bonus
df["salary_norm"] = df["salary"] / df["salary"].max()
df["bonus_norm"] = df["bonus_%"] / df["bonus_%"].max()

# Kombinerat score (viktat)
df["score"] = df["salary_norm"] + 2 * df["bonus_norm"]
````

Välj vilka kolumner som ska vara med i det nya DataFramet

Skapa två variabler top5 respektive bottom5 med enda skillnaden ascending=True/False

````python
columns = ["first_name", "salary", "bonus_%", "team", "score"]

df_top5 = df.nlargest(5, "score")[columns]
df_bottom5 = df.nsmallest(5, "score")[columns]
````
